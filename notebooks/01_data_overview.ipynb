```python
def churn_risk(row):
    if row['tenure'] < 12 and row['MonthlyCharges'] > 70:
        return 'High Risk'
    elif row['tenure'] < 24:
        return 'Medium Risk'
    else:
        return 'Low Risk'

df['ChurnRisk'] = df.apply(churn_risk, axis=1)
```
